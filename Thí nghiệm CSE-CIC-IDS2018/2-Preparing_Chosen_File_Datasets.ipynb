```python
import pandas as pd
```

### 1- Nối các bộ dữ liệu lại với nhau

```python
df_friday1 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Friday-02-03-2018_TrafficForML_CICFlowMeter.csv")
df_friday2 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Friday-16-02-2018_TrafficForML_CICFlowMeter.csv")
df_friday3 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Friday-23-02-2018_TrafficForML_CICFlowMeter.csv")
# df_thursday1 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Thuesday-20-02-2018_TrafficForML_CICFlowMeter.csv")
# df_thursday2 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Thursday-01-03-2018_TrafficForML_CICFlowMeter.csv")
df_thursday3 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Thursday-15-02-2018_TrafficForML_CICFlowMeter.csv")
df_thursday4 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Thursday-22-02-2018_TrafficForML_CICFlowMeter.csv")
df_wednesday1 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Wednesday-14-02-2018_TrafficForML_CICFlowMeter.csv")
df_wednesday2 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Wednesday-21-02-2018_TrafficForML_CICFlowMeter.csv")
# df_wednesday3 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Wednesday-28-02-2018_TrafficForML_CICFlowMeter.csv")
# df_tuesday1 = pd.read_csv("Dataset/CSE-CIC-IDS2018/Tuesday-20-02-2018_TrafficForML_CICFlowMeter_dropped_first_four_columns.csv")
```

```python
df_dataset = pd.concat([df_friday1, df_friday2, df_friday3, df_thursday3, df_thursday4, df_wednesday1, df_wednesday2], axis=0, ignore_index=True)
```

```python
# Hiển thị thông tin cơ bản về DataFrame
print("Số hàng và cột của df_dataset:", df_dataset.shape)
print("Các cột trong df_dataset:", df_dataset.columns.tolist())
print("Các nhãn duy nhất trong cột Label:", df_dataset["Label"].unique())
```

### 2- Loại bỏ dòng có nhãn 'Label' bị lỗi

```python
# Bỏ các dòng 'Label' lỗi của label
df_dataset.drop(df_dataset.loc[df_dataset["Label"] == "Label"].index, inplace=True)
```

```python
df_dataset
```

### 3- Tạo nhãn "Malicious" từ các nhãn "Bot", "DoS attacks-SlowHTTPTest" và "DoS attacks-Hulk"

```python
# Gộp các nhãn khác thành "Malicious", nhưng giữ lại "Benign"
df_dataset.replace(to_replace=['Bot', 'DoS attacks-SlowHTTPTest', 'DoS attacks-Hulk', 'Label', 'Brute Force -Web', 'Brute Force -XSS', 'SQL Injection', 'DoS attacks-GoldenEye', 'DoS attacks-Slowloris', 'FTP-BruteForce', 'SSH-Bruteforce', 'DDOS attack-LOIC-UDP', 'DDOS attack-HOIC'], value="Malicious", inplace=True)
```

```python
print("Các giá trị duy nhất trong cột Label:", df_dataset["Label"].unique())
```

```python
df_dataset
```

```python
df_dataset["Label"].unique()
```

```python
print(df_dataset["Label"].value_counts())
```

```python
print(df_dataset["Label"].value_counts()[['Benign']].sum())
print(df_dataset["Label"].value_counts()[['Malicious']].sum())
```

### 4- Checking "Protocol" values, and creating dummies based on the assumption that the Protocol values are categorical

```python
df_dataset["Protocol"].unique()
```

```python
# Ép buộc các nhãn trở thành kiểu chuỗi (string)
df_dataset = df_dataset.astype({"Protocol": str})
```

```python
df_dataset["Protocol"].unique()
```

```python
df_dataset = pd.get_dummies(df_dataset, columns=['Protocol'], drop_first=True)
```

```python
df_dataset
```

```python
# Đưa cột Label trở lại vị trí cuối cùng
df_dataset.insert(len(df_dataset.columns)-1, 'Label', df_dataset.pop('Label'))
```

```python
df_dataset
```

### 5- Loại bỏ một số cột có thể không liên quan (cần phân tích lại để xác định)

```python
columns_to_drop = [
    'Dst Port',
    'Timestamp',
    'Fwd PSH Flags',
    'Bwd PSH Flags',
    'Fwd URG Flags',
    'Bwd URG Flags',
    'Flow Byts/s',  # Trường này có chứa các giá trị np.inf trong quá trình huấn luyện, vì vậy đã bị loại bỏ.
    'Flow Pkts/s'  # Trường này có chứa các giá trị np.inf trong quá trình huấn luyện, vì vậy đã bị loại bỏ.
]
```

```python
df_dataset.drop(columns=columns_to_drop, inplace=True)
```

```python
df_dataset
```

### 6- Loại bỏ các giá trị bị thiếu hoặc trùng lặp

```python
df_dataset.dropna(inplace=True)
```

```python
df_dataset.drop_duplicates(inplace=True)
```

```python
df_dataset
```

### 7- Kiểm tra phân phối các nhãn sau khi loại bỏ các giá trị bị thiếu hoặc trùng lặp

* Malicious: 
* Benign: 

```python
print(df_dataset["Label"].value_counts()[['Benign']].sum())
print(df_dataset["Label"].value_counts()[['Malicious']].sum())
```

### 8- Thay thế nhãn Benign bằng số không và nhãn Malicious bằng số một 

```python
df_dataset.replace(to_replace="Benign", value=0, inplace=True)
df_dataset.replace(to_replace="Malicious", value=1, inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Khởi tạo biến đếm nhãn
label_counts = pd.Series(dtype=int)

# Đọc dữ liệu từng phần
chunk_size = 100000
for chunk in pd.read_csv('Dataset_To_Train_Models/processed_cse_cic_ids2018_malicious_benign.csv', chunksize=chunk_size):
    # Đếm số lượng nhãn trong mỗi phần
    chunk_counts = chunk['Label'].value_counts()
    label_counts = label_counts.add(chunk_counts, fill_value=0).astype(int)

# Vẽ biểu đồ cột
plt.figure(figsize=(8, 6))
label_counts.plot(kind='bar', color=['#4CAF50', '#FF5733'])
plt.title('Số lượng mẫu Benign và Malicious')
plt.xlabel('Nhãn')
plt.ylabel('Số lượng')
plt.xticks(rotation=0)
plt.grid(True, axis='y')
plt.show()

# In kết quả đếm
print("Số lượng từng nhãn:\n", label_counts)
```
